Problem 1 — Write Products into CSV File
Create products.csv and write header and product data using csv.writer().


```python
import csv

# Product data
products = [
    [101, "Laptop", "Electronics", 75000, 10],
    [102, "Mouse", "Accessories", 800, 50],
    [103, "Keyboard", "Accessories", 1500, 30]
]

# Create and write into CSV file
with open("products.csv", "w", newline="") as file:

    writer = csv.writer(file)

    # Header row
    writer.writerow(["id", "name", "category", "price", "quantity"])

    # Product rows
    writer.writerows(products)

print("Products CSV file created successfully!")
```

Problem 2 — Read Data from CSV File
Read all products from CSV file and display each row.


```python
import csv

# Open CSV file in read mode
with open("products.csv", "r") as file:

    reader = csv.reader(file)

    # Read and display each row
    for row in reader:
        print(row)
```

Problem 3 — Delete Product from CSV File
Delete product where id == 103 and rewrite updated data into CSV file


```python
import csv

updated_products = []

# Read existing CSV data
with open("products.csv", "r") as file:

    reader = csv.reader(file)

    # Store header
    header = next(reader)

    # Filter rows
    for row in reader:

        # Skip product with id 103
        if row[0] != "103":
            updated_products.append(row)

# Rewrite updated data into CSV
with open("products.csv", "w", newline="") as file:

    writer = csv.writer(file)

    # Write header
    writer.writerow(header)

    # Write updated rows
    writer.writerows(updated_products)

print("Product with ID 103 deleted successfully!")
```

Problem 4 — Display Products with Price Greater than 5000
Display all products whose price is greater than 5000.


```python
import csv

# Read CSV file
with open("products.csv", "r") as file:

    reader = csv.reader(file)

    # Skip header
    header = next(reader)

    print("Products with price greater than 5000:\n")

    # Check each product
    for row in reader:

        price = int(row[3])

        if price > 5000:
            print(row)
```

Problem 5 — Write Products into JSON File
Create products.json and write product data using json.dump().


```python
import json

# Product data
products = [
    {"id": 101, "name": "Laptop", "category": "Electronics", "price": 75000, "quantity": 10},
    {"id": 102, "name": "Mouse", "category": "Accessories", "price": 800, "quantity": 50},
    {"id": 103, "name": "Keyboard", "category": "Accessories", "price": 1500, "quantity": 30},
    {"id": 104, "name": "Mobile", "category": "Electronics", "price": 25000, "quantity": 15},
    {"id": 105, "name": "Headphones", "category": "Accessories", "price": 2000, "quantity": 20}
]

# Write data into JSON file
with open("products.json", "w") as file:

    json.dump(products, file, indent=4)

print("Products JSON file created successfully!")
```

Problem 6 — Read Data from JSON File
Read and display all products from JSON file.


```python
import json

# Open JSON file
with open("products.json", "r") as file:

    # Load JSON data into Python
    products = json.load(file)

    # Display products
    for product in products:
        print(product)
```

Problem 7 — Delete Product from JSON File
Delete product where id == 102.


```python
import json

# Read existing JSON data
with open("products.json", "r") as file:

    products = json.load(file)

# Remove product with id 102
updated_products = []

for product in products:

    if product["id"] != 102:
        updated_products.append(product)

# Rewrite updated data into JSON file
with open("products.json", "w") as file:

    json.dump(updated_products, file, indent=4)

print("Product with ID 102 deleted successfully!")
```

Problem 8 — Update Product Quantity
Update quantity of Laptop to 25.


```python
import json

# Read JSON data
with open("products.json", "r") as file:

    products = json.load(file)

# Update Laptop quantity
for product in products:

    if product["name"] == "Laptop":
        product["quantity"] = 25

# Rewrite updated data into JSON file
with open("products.json", "w") as file:

    json.dump(products, file, indent=4)

print("Laptop quantity updated successfully!")
```

Problem 10 — Append New Product to CSV
Append a new Monitor product into CSV file.


```python
import csv

# New product
new_product = [106, "Monitor", "Electronics", 12000, 12]

# Open CSV file in append mode
with open("products.csv", "a", newline="") as file:

    writer = csv.writer(file)

    # Append new row
    writer.writerow(new_product)

print("New product added successfully!")
```

Problem 11 — Append New Product to JSON
Append the same product into JSON file.




```python
import json

# New product
new_product = {
    "id": 106,
    "name": "Monitor",
    "category": "Electronics",
    "price": 12000,
    "quantity": 12
}

# Read existing JSON data
with open("products.json", "r") as file:

    products = json.load(file)

# Append new product
products.append(new_product)

# Rewrite updated data into JSON file
with open("products.json", "w") as file:

    json.dump(products, file, indent=4)

print("New product added to JSON successfully!")
```
