# 1. Array Sum
Write a function that accepts a list of numbers and returns the sum of all elements.
numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
# Function for array sum
def arraysum(arr):

# Starting total = 0
    total = 0

    # Loop through each number
    for num in arr:

        # Add current number into total
        total += num

    # Return final total
    return total


arraysum(numbers)
```

# 2. Maximum Element
Write a function to find the largest element in a list without using max().numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def find_max(arr):

    largest = arr[0]

    for num in arr:

        if num > largest:
            largest = num

    return largest


print("Maximum Element:", find_max(numbers))
```

# 3. Even Numbers Counter
Write a function that counts how many even numbers are present in a list.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def count_even(arr):

    count = 0

    for num in arr:

        if num % 2 == 0:
            count += 1

    return count


print("Even Numbers Count:", count_even(numbers))
```

# 4. Reverse Array
Write a function to reverse the elements of a list.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def reverse_array(arr):

    reversed_list = arr[::-1]

    return reversed_list


print("Reversed Array:", reverse_array(numbers))
```

# 5. Search Element
Write a function to search for a given element in a list and return its index.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def search_element(arr, target):

    for index in range(len(arr)):

        if arr[index] == target:
            return index

    return -1


print("Element Found At Index:", search_element(numbers, 90))
```

# 6. Palindrome List
Write a function to check whether a list reads the same forward and backward.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]


```python
def is_palindrome(arr):

    if arr == arr[::-1]:
        return True

    return False


print("Palindrome Check:", is_palindrome(numbers))
```

# 7. Second Largest Number
Write a function to find the second largest number in a list.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def second_largest(arr):

    unique_numbers = list(set(arr))

    unique_numbers.sort()

    return unique_numbers[-2]


print("Second Largest Number:", second_largest(numbers))
```

# 8. Duplicate Elements
Write a function that identifies duplicate elements in a list.numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
def find_duplicates(arr):

    duplicates = []

    seen = []

    for num in arr:

        if num in seen and num not in duplicates:
            duplicates.append(num)

        seen.append(num)

    return duplicates


print("Duplicate Elements:", find_duplicates(numbers))
```

# 9. Function for Student Marks
Create a function that accepts student marks in a list and returns average, highest, and lowest marks. numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

```python
student_marks = [78, 85, 90, 67, 88]

def marks_analysis(marks):

    average = sum(marks) / len(marks)

    highest = max(marks)

    lowest = min(marks)

    return average, highest, lowest


avg, high, low = marks_analysis(student_marks)

print("Average Marks:", avg)
print("Highest Marks:", high)
print("Lowest Marks:", low)
```
