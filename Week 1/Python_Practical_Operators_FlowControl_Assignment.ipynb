# Q5. Given total items = 37 and box capacity = 6, calculate how many full boxes are required and how many items remain.

```python

items = 37
capacity = 6

full_boxes = items // capacity

remaining_items = items % capacity

print("Full Boxes-", full_boxes)
print("Remaining Items-", remaining_items)
```

# Q11. Iterate through a list of orders and count how many orders are above 1,00,000.

```python

orders = [50000, 150000, 250000, 70000]
count = 0

for order in orders:
    if order > 100000:
        count += 1

print("Total Orders Above 100000:", count)
```

# Q12. Simulate payment retry logic where the system retries payment a maximum of 3 times if payment fails.

```python

attempt = 1
while attempt <= 3:

    print("Payment Failed: Retry", attempt)

    attempt += 1
```

# Q13. Write a program that keeps checking order status until status becomes 'Delivered' or maximum 5 attempts are reached.

```python

status = "Pending"
attempt = 1
while status != "Delivered" and attempt <= 5:

    print("Checking Order Status As Attempt", attempt)

    attempt += 1

print("Process Finished")
```

# Q14. Inventory restock check: Keep checking stock level every iteration and stop when stock becomes greater than 100.

```python

stock = 20
while stock <= 100:
    print("Current Stock:", stock)
    # Restock
    stock += 20
print("Stock Sufficient")
```

# Q15. You are given a list of orders with order_id and amount.
- Skip cancelled orders
- Classify remaining orders as Premium or Regular
- Count total premium orders

```python
orders = [
    {"id": 101, "amount": 120000, "status": "Completed"},
    {"id": 102, "amount": 40000, "status": "Cancelled"},
    {"id": 103, "amount": 75000, "status": "Cancelled"},
    {"id": 104, "amount": 150000, "status": "Completed"}
]

premium_count = 0

# Loop through orders
for order in orders:

    # Skip cancelled orders
    if order["status"] == "Cancelled":
        continue

    # Premium order checking
    if order["amount"] > 100000:

        print(order["id"],"- Premium Order")

        premium_count += 1

    else:
        print(order["id"],"- Regular Order")

# Final premium count
print("Total Premium Orders:", premium_count)
```
