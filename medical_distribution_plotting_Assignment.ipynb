## 1. Import Libraries and Create Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating Dataset
medical_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Medicine_Distributed': [1200, 1500, 1800, 1700, 2100, 2500],
    'Hospital_Orders': [30, 45, 50, 48, 60, 75],
    'Delivery_Time': [5, 4, 6, 5, 3, 2],
    'Revenue': [50000, 65000, 72000, 70000, 85000, 95000]
}

df = pd.DataFrame(medical_data)

print(df)
```

# LINE CHART PROBLEMS
## 1. Medicine Distribution Over Months

```python
plt.figure(figsize=(8,5))

plt.plot(df['Month'],
         df['Medicine_Distributed'],
         marker='o')

plt.title("Medicine Distribution Over Months")
plt.xlabel("Month")
plt.ylabel("Medicine Distributed")
plt.grid(True)

plt.savefig("line_medicine_distribution.png")
plt.show()
```

## 2. Monthly Revenue Growth

```python
plt.figure(figsize=(8,5))

plt.plot(df['Month'],
         df['Revenue'],
         marker='s')

plt.title("Monthly Revenue Growth")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.savefig("line_revenue_growth.png")
plt.show()
```

## 3. Multiple Line Chart

```python
plt.figure(figsize=(8,5))

plt.plot(df['Month'],
         df['Medicine_Distributed'],
         marker='o',
         label='Medicine Distributed')

plt.plot(df['Month'],
         df['Hospital_Orders'],
         marker='s',
         label='Hospital Orders')

plt.title("Medicine Distribution vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.grid(True)

plt.savefig("multiple_line_chart.png")
plt.show()
```

# BAR CHART PROBLEMS
## 4. Vertical Bar Chart

```python
plt.figure(figsize=(8,5))

plt.bar(df['Month'],
        df['Hospital_Orders'])

plt.title("Hospital Orders by Month")
plt.xlabel("Month")
plt.ylabel("Orders")

plt.savefig("vertical_bar_chart.png")
plt.show()
```

## 5. Horizontal Bar Chart

```python
plt.figure(figsize=(8,5))

plt.barh(df['Month'],
         df['Revenue'])

plt.title("Revenue by Month")
plt.xlabel("Revenue")
plt.ylabel("Month")

plt.savefig("horizontal_bar_chart.png")
plt.show()
```

## 6. Grouped Bar Chart

```python
x = np.arange(len(df['Month']))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x - width/2,
        df['Medicine_Distributed'],
        width,
        label='Medicine Distributed')

plt.bar(x + width/2,
        df['Hospital_Orders'],
        width,
        label='Hospital Orders')

plt.xticks(x, df['Month'])

plt.title("Medicine Distributed vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()

plt.savefig("grouped_bar_chart.png")
plt.show()
```

# SCATTER PLOT PROBLEMS
## 7. Hospital Orders vs Revenue

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Hospital_Orders'],
            df['Revenue'])

plt.title("Hospital Orders vs Revenue")
plt.xlabel("Hospital Orders")
plt.ylabel("Revenue")

plt.savefig("scatter_orders_revenue.png")
plt.show()
```

## 8. Delivery Time vs Revenue

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Delivery_Time'],
            df['Revenue'])

plt.title("Delivery Time vs Revenue")
plt.xlabel("Delivery Time")
plt.ylabel("Revenue")

plt.savefig("scatter_delivery_revenue.png")
plt.show()
```

## 9. Medicine Distributed vs Hospital Orders

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Medicine_Distributed'],
            df['Hospital_Orders'])

plt.title("Medicine Distributed vs Hospital Orders")
plt.xlabel("Medicine Distributed")
plt.ylabel("Hospital Orders")

plt.savefig("scatter_medicine_orders.png")
plt.show()
```

# PIE CHART PROBLEMS
## 10. Medicine Distribution Percentage by Month

```python
plt.figure(figsize=(8,5))

plt.pie(df['Medicine_Distributed'],
        labels=df['Month'],
        autopct='%1.1f%%')

plt.title("Medicine Distribution Percentage")

plt.savefig("pie_medicine_distribution.png")
plt.show()
```

# 11. Revenue Contribution by Month

```python
plt.figure(figsize=(8,5))

plt.pie(df['Revenue'],
        labels=df['Month'],
        autopct='%1.1f%%')

plt.title("Revenue Contribution by Month")

plt.savefig("pie_revenue_contribution.png")
plt.show()
```

## 12. Highest Revenue Month Using Explode

```python
explode = [0,0,0,0,0,0.2]   # June has highest revenue

plt.figure(figsize=(8,5))

plt.pie(df['Revenue'],
        labels=df['Month'],
        autopct='%1.1f%%',
        explode=explode)

plt.title("Highest Revenue Month Highlighted")

plt.savefig("pie_highest_revenue.png")
plt.show()
```
